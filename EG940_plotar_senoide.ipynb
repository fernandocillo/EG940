```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

$$x[n] = x(nT_S)$$
$$x(t)=cos(2\pi f_0t)$$
$$x[n]=cos(2\pi f_0nT_S)$$

```python
fs=44100
Ts=1/fs
f0=440
n=np.arange(1000) #n√∫mero de amostras
xn=np.cos(2*np.pi*f0*n*Ts)
```

```python
plt.figure()
plt.plot(n[40:200]*Ts, xn[40:200], label='Sinal amostrado')
plt.ylabel('Magnitude')
plt.xlabel('Tempo(s)')
plt.legend()
plt.plot()
```

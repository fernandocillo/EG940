```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

$$x[n] = x(nT_S)$$
$$x(t)=cos(2\pi f_0t)$$
$$x[n]=cos(2\pi f_0nT_S)$$

```python
fs=44100
Ts=1/fs
f0=440
n=np.arange(4410*5) #n√∫mero de amostras
xn = np.cos(2 * np.pi * f0 * n * Ts)

#x_random = np.random.randn(len(n))
```

```python
plt.figure()
plt.plot(n*Ts, xn, label='Sinal amostrado')
plt.ylabel('Tempo(s)')
plt.show()
```

```python
xk = np.abs(np.fft.fft(xn))
freqs=np.arange(22050)
freqs=freqs* fs/ 22050

plt.figure()
plt.plot(freqs,xk)
plt.xlim([400,500]) #valor esperado: pico em 440 Hz
plt.show()

plt.figure()
plt.plot(freqs,xk)
plt.xlim([44100-500,44100-400]) #f√™nomeno de aliasing
plt.show()

#a metade superior da amostragem √© o reflexo da metade inferior, ent√£o n√£o precisa ser mostrada
```

```python
xk = np.abs(np.fft.rfft(xn))
freqs=np.arange(len(xk))
freqs=freqs* 0.5*fs/len(freqs)

plt.figure()
plt.plot(freqs,xk)
#plt.xlim([400,500]) #pico
plt.show()
```

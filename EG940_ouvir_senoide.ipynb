```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import Audio
```

$$x[n] = x(nT_S)$$
$$x(t)=cos(2\pi f_0t)$$
$$x[n]=cos(2\pi f_0nT_S)$$

```python
fs=44100
Ts=1/fs
f0=440
n=np.arange(44100*5) #n√∫mero de amostras

x1n = np.cos(2 * np.pi * f0 * n * Ts)* 0.1
x2n = np.cos(2 * np.pi * 2*f0 * n * Ts)* 0.1
xn = x1n + x2n

#x_random = np.random.randn(len(n)) * 0.1
```

```python
Audio(data=xn, rate=fs)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fs = 40000
Ts= 1/fs
f0 = 400
n = np.arange(400)
xn = np.cos(2 * np.pi * f0 * n * Ts)
```

```python
## SEM JANELAMENTO -> vazamento de frequência

xk = 20*np.log10(np.abs(np.fft.rfft(xn, len(xn)*10))) #aumenta resolução
xk -= np.max(xk)
freqs = np.arange(len(xk))
freqs = freqs * fs * 0.5/len(freqs)
plt.figure()
plt.plot(freqs, xk)
plt.xlim([000, 800])
plt.ylim([-50, 0])
plt.show()
```

```python
# JANELA

w = np.hanning(400)
plt.figure()
plt.plot(w)
plt.show()
```

```python
## COM JANELAMENTO -> elimina o fenômeno de Gibbs

w=np.hanning(len(xn))

xk = 20*np.log10(np.abs(np.fft.rfft(xn*w, len(xn)*10))) #aumenta resolução
xk -= np.max(xk)
freqs = np.arange(len(xk))
freqs = freqs * fs * 0.5/len(freqs)
plt.figure()
plt.plot(freqs, xk)
plt.xlim([000, 800])
plt.ylim([-50, 0])
plt.show()
```

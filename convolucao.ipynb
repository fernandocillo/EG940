**Convolução:**
$$ a[n]*x[n] = ∑_{m=0}^{M-1}a[m]x[n-m]$$

$$h[n]*δ[n]=h[n]$$
em que $h[n]$ é a resposta ao impulso e $\delta[n]$ é o elemento neutro da convolução

Domínio do tempo: $y[n]=h[n]*x[n]$ \\

Domínio da frequência: $Y[k]=H[k]\cdot X[k]$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = np.ones(5)
x = np.random.randn(300)/10
hx = np.convolve(h, x)

H = np.abs(np.fft.rfft(h, 500))
X = np.abs(np.fft.rfft(x, 500))
HX = np.abs(np.fft.rfft(hx, 500))

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(H, label='H[k]')
plt.plot(X, label='X[k]')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(HX, label='HX[k]')
plt.legend()
plt.show()
```

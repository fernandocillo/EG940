```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 5, 6, 0, 6, 3])
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
Y = np.fft.fft(y) #biblioteca numpy -> pacote fft -> função fft
#print(Y)
```

```python
plt.figure()
plt.plot(np.abs(Y))
plt.show()

plt.figure()
plt.plot(np.angle(Y))
plt.show()
```

```python
y2 = np.fft.ifft(Y) #transformada inversa

plt.figure()
plt.plot(x,y, linewidth=5, label='Sinal original')
plt.plot(x,abs(y2), linewidth=2, label='FFT + IFFT')
plt.legend()
plt.show()
```
